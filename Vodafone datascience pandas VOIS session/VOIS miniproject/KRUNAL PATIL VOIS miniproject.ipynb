### Task 3: Find out the per unit price from the data

###### a. Convert date time field into pandas time object

```python
store_data["Order Date"] = pd.to_datetime(store_data["Order Date"])
store_data["Ship Date"] = pd.to_datetime(store_data["Ship Date"])

store_data[['Order Date','Ship Date']].info()   # here the date is converted to datetime64 object
```

###### b. Find out the price per unit from the data and create a column for it in the same data

```python
store_data["Price per unit"] = store_data["Sales"] / store_data["Quantity"]
store_data.head()
```

### Task 4: Find out the monthly revenue and analyze the findings

```python
monthly_revenue = pd.DataFrame(store_data, columns = ['Ship Date'])
monthly_revenue.rename(columns={'Ship Date': 'Date'}, inplace=True)
monthly_revenue.sort_values(by = ['Date'], inplace=True)
monthly_revenue
```

###### a. Create a column with Year and month field only give a suitable name to it

```python
monthly_revenue['Month'] = monthly_revenue['Date'].dt.strftime('%Y-%m')
monthly_revenue.drop(['Date'], axis=1, inplace=True)
monthly_revenue.drop_duplicates(inplace=True)
monthly_revenue
```

###### b.Create a separate dataset from the data which will have two columns , one which is created in step a of this task and second the monthly sales / revenue

```python
monthly_sales = monthly_revenue.groupby(monthly_revenue['Month'])['Sales'].sum()
monthly_sales = pd.DataFrame(monthly_sales)
monthly_sales
```

###### c. Plot this dataset

```python
monthly_sales.plot(y='Sales', kind = 'line')
plot.show()
```

### Task 5:  Find out the yearly revenue and analyze the findings

###### a. Create a column with Year field only give a suitable name to it

```python
yearly_revenue = pd.DataFrame(store_data, columns = ['Ship Date'])
yearly_revenue.rename(columns={'Ship Date': 'Date'}, inplace=True)
yearly_revenue.sort_values(by = ['Date'], inplace=True)

yearly_revenue['Year'] = yearly_revenue['Date'].dt.strftime('%Y')
yearly_revenue.drop(['Date'], axis=1, inplace=True)

yearly_revenue
```

###### b. Create a separate dataset from the data which will have two columns , one which is created in step a of this task and second the yearly sales / revenue

```python
yearly_revenue['Sales'] = store_data['Sales'].astype(float)
yearly_sales = yearly_revenue.groupby(yearly_revenue['Year'])['Sales'].sum()
yearly_sales = pd.DataFrame(yearly_sales)
yearly_sales
```

###### c. Plot this dataset

```python
yearly_sales.plot(y='Sales', kind = 'bar')
plot.show()
```

### Task 6: Finding out the monthly growth rate and analyse the findings

###### a. Create a column in the monthly revenue dataset for monthly growth rate

```python
monthly_sales['Growth'] = monthly_sales['Sales'].pct_change()
monthly_sales
```

###### b. Find out the maximum monthly growth rate

```python
monthly_sales.sort_values('Growth', ascending=False).head(1)
```

###### c. Plot this dataset

```python
monthly_sales.plot(y='Growth', title="Growth rate", kind = 'bar');
plot.show();
```

### Task 7: Finding out the most and least sold product id

###### a. Create a new dataset including the product ids and total quantities sold for that id

```python
product = pd.DataFrame(store_data, columns = ['Row ID', 'Product ID'])
product['Quantities sold'] = product.groupby(['Product ID']).transform('count')
product.drop(['Row ID'], axis=1, inplace=True)
product.drop_duplicates(inplace=True)
product = product.reset_index(drop=True)
product
```

###### b. Find out the most sold product from the created dataset

```python
store_data.loc[ store_data[ 'Product ID'] == product.max()['Product ID'] ]['Product Name'].unique()[0]
```

###### c. Find out the least sold product from the created dataset

```python
store_data.loc[ store_data[ 'Product ID'] == product.min()['Product ID'] ]['Product Name'].unique()[0]
```

### Task 10: Finding out the majority and minority customer cities on basis of
